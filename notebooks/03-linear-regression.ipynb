## Linear regression 

**What we will learn?**
- Linear regression 
- Handling of missing data 
- Training and test set splitting 
- Linear regression with multiple predictor variable

```python
# Import required packages 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno
import seaborn as sns

%matplotlib inline
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
# Load car data set
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",\
          "num-of-doors", "body-style", "drive-wheels", "engine-location",\
          "wheel-base", "length", "width", "height", "curb-weight", "engine-type",\
          "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",\
          "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg",\
          "price"]
data = pd.read_csv(url, header=None, names=headers, na_values=["?"])
```

```python
# Visual inspection of missing values
missingno.matrix(data, figsize = (30,5))
plt.show()
```

```python
data.dropna(subset=["num-of-doors", "bore", "stroke", "horsepower", \
                    "peak-rpm", "price"], axis=0, inplace=True)
```

```python
# We can drop normalized losses column to simplify our analysis
# data.drop(columns=["normalized-losses"], inplace=True)

# Alternatively, we could insert the missing values with its mean
mean = data["normalized-losses"].mean()
data["normalized-losses"].replace(np.nan, mean, inplace=True)
```

```python
sns.regplot(x="engine-size", y="price", data=data)
plt.show()
```

```python
# Residual plot
sns.residplot(x="engine-size", y="price", data=data)
plt.show()
```

```python
lm = LinearRegression()
X = data[["engine-size"]]
Y = data[["price"]]
lm.fit(X, Y);
print(lm.intercept_, lm.coef_)
```

```python
lm.predict(np.array([[100]]))
```

### Train-test split

```python
X, y = data[["engine-size"]], data[["price"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, \
                                                    random_state=0)
lm.fit(X_train, y_train)
lm.score(X_test, y_test)
```

```python
# training score
lm.score(X_train, y_train)
```

### Linear regression with multiple predictor variable

```python
X, y = data[['horsepower', 'highway-mpg']], data[["price"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, \
                                                    random_state=0)
lm.fit(X_train, y_train)
lm.score(X_test, y_test)
```
