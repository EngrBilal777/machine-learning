Here we will apply Hierarchical clustering to animals with attributes dataset. There is a list of 50 animals with 85 features. Original dataset can be found at - <https://cvml.ist.ac.at/AwA2/>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
%matplotlib inline

plt.rcParams["figure.dpi"] = 100
plt.rcParams["figure.facecolor"] = "white"
```

```python
# Load in animal names
f = open("../datasets/Animals_with_Attributes2/classes.txt")
lines = f.readlines()
names = []
for line in lines:
    (key,val) = line.split()
    names.append(val)
f.close()

# Load feature matrix
data = np.loadtxt("../datasets/Animals_with_Attributes2/predicate-matrix-continuous.txt")
```

```python
model = sch.linkage(data, method='ward')
```

```python
plt.rcParams["figure.figsize"] = (5, 10)
plt.xticks([])
dend = sch.dendrogram(model, orientation='left', labels=names, leaf_font_size=10)
```
