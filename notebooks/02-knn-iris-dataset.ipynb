## K-nearest neighbors for classification 

**What we will learn?** 
- K-nearest neighbors algorithm for classification of Iris flower species 

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Import Iris dataset from scikit learn
data = load_iris()
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
pd.plotting.scatter_matrix(df, c=data.target, figsize=(15, 15), marker='o',\
                 hist_kwds={'bins': 20}, s=60, alpha=.8, cmap='Accent')
plt.show()
```

### Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(data['data'], \
                                   data['target'], random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.predict(X_test[0].reshape(1, -1))[0]
```

```python
y_test[0]
```

```python
y_pred = knn.predict(X_test)
np.mean(y_pred == y_test)
```

```python
knn.score(X_test, y_test)
```

```python
knn2 = KNeighborsClassifier(n_neighbors=30)
knn2.fit(X_train, y_train)
knn2.score(X_test, y_test)
```
